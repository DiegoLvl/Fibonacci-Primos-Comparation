##**<center>"Now we import the libraries</center>**

```python
import pandas as pd
import matplotlib as mt
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
#Quantity to compare
z = 15
```

##**<center>Dataframe of Fibonacci numbers</center>**




```python
def fibonacci(n):
    fib = [0,1]
    
    while len(fib) < n:
     fib.append(fib[-1] + fib[-2])

    return fib

numeros_fibonacci = fibonacci (z)

dfy=pd.DataFrame({'fibonacci': numeros_fibonacci})
y = dfy['fibonacci']
```

##**<center>Dataframe of prime numbers</center>**

```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


numeros_primos = []
numero = 2
while len(numeros_primos) < z:
    if es_primo(numero):
        numeros_primos.append(numero)
    numero += 1

dfx=pd.DataFrame({'dataprimos': numeros_primos})
x = dfx['dataprimos']
print (x, y)
```

#**<center>Let's compare the first eleven numbers</center>**
>
Let's start with bars.



```python
indices = np.arange(len(x))
width = 0.25
fig, ax = plt.subplots(figsize=(10, 6))


rects_y = ax.bar(indices, y, width, label='Fibonacci')
rects_x = ax.bar(indices + width, x, width, color="red", label='Primos')

ax.set_title('Comparaci√≥n Fibonacci vs Primos')
ax.set_xticks(indices)
ax.set_xticklabels([])
ax.legend()


plt.show()
```

##<center>We can observe that the Fibonacci sequence diverges from the sequence of prime numbers, making it impossible to find a pattern.</center>
